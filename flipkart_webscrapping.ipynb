```python
import requests #make a request to web page
import bs4 # pulling out data from html and xml
```

```python
request1= requests.get("https://www.flipkart.com/oppo-f19-prism-black-128-gb/product-reviews/itm82f60569dc6db?pid=MOBGFGWNUZXAM7QY&lid=LSTMOBGFGWNUZXAM7QYB6Z1TY&marketplace=FLIPKART")
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

```python
reviews=soup.findAll('div',{'class': 't-ZTKy'})
for review in reviews:
    print(review.get_text()+"\n\n")
```
